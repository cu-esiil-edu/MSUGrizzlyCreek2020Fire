```python
import pandas as pd

# For deeper analysis
import numpy as np  # for adding trendline to plot
import matplotlib.pyplot as plt  # for plotting
```

```python
# Downstream of fire of boundary, Glenwood Springs
ncei_weather_url = (
    'https://www.ncei.noaa.gov/access/services/data/v1'
    '?dataset=daily-summaries'
    '&dataTypes=TOBS,PRCP'
    '&stations=US1COGF0060'  # A glenwood springs weather station, 99% coverage https://www.ncei.noaa.gov/cdo-web/results
    '&startDate=1949-10-01'
    '&endDate=2024-02-18'
    '&includeStationName=true'
    '&includeStationLocation=1'
    '&units=standard')
ncei_weather_url
```

```python
downstream_w_df = pd.read_csv(
    ncei_weather_url,
    index_col="DATE",
    parse_dates=True,
    na_values=["NaN"]
)
downstream_w_df
```

```python
# Upstream of fire boundary, Steamboat Springs
ncei_weather_url2 = (
    'https://www.ncei.noaa.gov/access/services/data/v1'
    '?dataset=daily-summaries'
    '&dataTypes=TOBS,PRCP'
    '&stations=US1CORT0016'
    '&startDate=1949-10-01'
    '&endDate=2024-02-18'
    '&includeStationName=true'
    '&includeStationLocation=1'
    '&units=standard')
ncei_weather_url2
```

```python
upstream_w_df = pd.read_csv(
    ncei_weather_url2,
    index_col="DATE",
    parse_dates=True,
    na_values=["NaN"]
)
upstream_w_df
```

```python
# Steamboat springs station

# Making dataframe only precipitation
upstream_w_df = upstream_w_df[['PRCP']]

# Replacing any NaN values with zero
upstream_w_df['PRCP'] = upstream_w_df['PRCP'].replace(np.nan, 0)

# Removing all rows that have a precipitation of zero
upstream_w_df = upstream_w_df.loc[(upstream_w_df!=0).any(axis=1)]
upstream_w_df
```

```python
# Glenwood Springs station

# Making dataframe only precipitation
downstream_w_df = downstream_w_df[['PRCP']]

# Replacing any NaN values with zero
downstream_w_df['PRCP'] = downstream_w_df['PRCP'].replace(np.nan, 0)

# Removing all rows that have a precipitation of zero
downstream_w_df = downstream_w_df.loc[(downstream_w_df!=0).any(axis=1)]
downstream_w_df
```
